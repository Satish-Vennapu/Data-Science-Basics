## Data Visualization with Matplotlib

### Line Graph

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.plot([1,2,3,5]);
```

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.figure(figsize=(12, 6));
sns.set_style('darkgrid')
plt.plot(years,apples,'s-b')
plt.plot(years,oranges, marker='o',c='r',lw=2,ls='-',ms=8,mew=2,mec='green', alpha=0.5)
# color,linewidth,linestyle,markerstyle, markeredgewidth, markeredgecolor,alpha=opacity(0-1)
plt.xlabel('Years',fontsize=14)
plt.ylabel('Yield (Tones per acre)',fontsize=14)
plt.title('Crop Yields',fontsize=24)
plt.legend(['Apples', 'Oranges']);
plt.tick_params(axis='both',which='major',labelsize=14)
```

```python
# edit default styles directly by modifying the matplotlib.rcParams dictiona
import matplotlib
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']=(10,6)
matplotlib.rcParams['figure.facecolor']='#00000000'
```

### Scatter Plot

```python
x_values=list(range(1,1001))
y_values=[x**2 for x in x_values]
plt.scatter(x_values,y_values,s=40,edgecolors='None',c=(0.1,0.8,0.8)) 
# s= size, edgecolor of dot,(R,G,B)
plt.axis([0,1100,1,1100000])
```

```python
plt.scatter(x_values,y_values,s=20,c=y_values,cmap=plt.cm.Reds)
plt.savefig('squares_plot.png',bbox_inches='tight')
```

```python
flowers_df=sns.load_dataset('iris')
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length,flowers_df.sepal_width);
```

```python
plt.scatter(flowers_df.sepal_length,flowers_df.sepal_width,s=100,edgecolors='k') # s=size of dot
plt.axis([4,8,2,4.5]) # min,max of x and y-axis
```

```python
plt.figure(figsize=(12,6))
plt.title('Sepal length vs width')
sns.scatterplot(x=flowers_df.sepal_length,y=flowers_df.sepal_width, hue=flowers_df.species,s=100);
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',s=100,data=flowers_df);
```

### Histogram

```python
plt.hist(flowers_df.sepal_width)
plt.title('Distribution of Sepal width');
```

```python
#bins = intervals
import numpy as np
plt.hist(flowers_df.sepal_width, color='green',bins=np.arange(2,5,0.25));
```

```python
#unuequal sized bins
plt.hist(flowers_df.sepal_width,color='purple',bins=[1,3,3.5,4,4.5]);
```

```python
#multiple histograms
setosa_df=flowers_df[flowers_df.species=='setosa']
virginica_df=flowers_df[flowers_df.species=='virginica']
versicolor_df=flowers_df[flowers_df.species=='versicolor']
```

```python
plt.hist(setosa_df.sepal_width,bins=np.arange(2,5,0.25),alpha=0.5)
plt.hist(virginica_df.sepal_width,bins=np.arange(2,5,0.25),alpha=0.5);
```

```python
# Stacking histograms
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],bins=np.arange(2,5,0.25),stacked=True)
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
plt.title("Distribution of Sepal_widths");
```

### Bar chart

```python
# Stacking bars
plt.bar(years,apples)
plt.bar(years, oranges,bottom=apples)
plt.legend(['apples','oranges']);
```

### Heat Map

```python
flights_df
```

```python
sns.heatmap(flights_df,annot=True,fmt='d',cmap='Blues')
```

### Multiple Charts in a grid

```python
fig,axes=plt.subplots(2,2,figsize=(16,8))
#Passing the axes into seaborn
axes[0,0].set_title('Sepal_length vs Sepal_width')
sns.scatterplot(x=flowers_df.sepal_length,y=flowers_df.sepal_width,hue=flowers_df.species,s=100,ax=axes[0,0]);

# Using the axes for plotting
axes[0,1].set_title('Distribution of Sepal Width')
axes[0,1].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],bins=np.arange(2, 5, 0.25),stacked=True);
axes[0,1].legend(['Setosa','Versicolor','Virginica'])

axes[1,0].set_title('Restaurant bills')
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, ax=axes[1,0]);

axes[1,1].set_title('Flight traffic')
sns.heatmap(flights_df, cmap='Blues', ax=axes[1,1]);
```
